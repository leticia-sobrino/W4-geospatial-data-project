### 1. Busco aquellas empresas que estén ubicadas en Madrid y londres

```python
collection.find_one({})
```

```python
collection.find_one({}).keys()
```

* Para ver en donde se encuentran las empresas tengo que meterme dentro de offices y a su vez buscar el "city" donde están para poder diltrar por Madrid y Londres

```python
collection.find_one({"name": "Google"}, {"name": 1, "category_code":1, "_id": 0})
```

```python
madrid_companies = collection.find({"offices.city": "Madrid", "category_code" : "search"}, {"name": 1,"category_code":1, "_id" : 0})
```

```python
list(madrid_companies)
```

```python
london_companies = collection.find({"offices.city": "London"}, {"name": 1, "_id" : 0})
```

```python
list(london_companies)
```
